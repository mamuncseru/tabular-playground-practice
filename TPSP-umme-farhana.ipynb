# Tabular Playground Series - January - Kaggle 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
train_data.shape
```

```python
test_data.shape
```

```python
train_data.head()
```

```python
test_data.head()
```

## Data preprocessing

```python
train_data = train_data.drop(['id'], axis=1)
test = test_data.drop(['id'], axis=1)
```

```python
train_data.head()
```

```python
test.head()
```

## Split data x_train and y_train

```python
x_train = train_data.drop(['target'], axis=1)
y_train = train_data['target']
```

```python
x_train.head()
```

```python
y_train.head()
```

## Build a Model By Linear Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
l_regressor = LinearRegression().fit(x_train, y_train)
```

```python
l_regressor.score(x_train, y_train)
```

```python
y_predict_lin = l_regressor.predict(test)
print(y_predict_lin)
```

```python
submission = pd.DataFrame({
    "id": test_data["id"],
    "target": y_predict_lin
})

submission.to_csv('submission.csv', index=False)
```

## Build a Model By Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
d_regressor = DecisionTreeRegressor(random_state=0)
d_regressor.fit(x_train, y_train)
```

```python
y_predict_dec = d_regressor.predict(test)
print(y_predict_dec)
```

```python
submission = pd.DataFrame({
    "id": test_data["id"],
    "target": y_predict_dec
})

submission.to_csv('submission_01.csv', index=False)
```
