```python
data = pd.read_csv('../input/titanic/train.csv')
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Fare', 'SibSp', 'Parch'], axis=1, inplace=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])

data.head()
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
Y_train = data['Survived']
X_train = data.drop('Survived', axis=1)
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
seed = np.random.seed(0)
```

```python
kfold = KFold(n_splits=10, random_state=seed)
dtc = DecisionTreeClassifier()
```

```python
num_trees = 160
```

```python
model = BaggingClassifier(base_estimator=dtc, n_estimators=num_trees, random_state=seed)
```

```python
results = cross_val_score(model, X_train, Y_train, cv=kfold)
print(results.mean())
```

```python
model2 = RandomForestClassifier(n_estimators=num_trees, max_features=4)
results2 = cross_val_score(model2, X_train, Y_train, cv=kfold)
print(results2.mean())
```

```python
model.score(X_train, Y_train)
```
